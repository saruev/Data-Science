### Problem 3. Column manipulation (1 point)
Make the column names more Pythonic (which helps with the quality and... aesthetics). Convert column names to `snake_case`, i.e. `species`, `country_of_origin`, `ico_number`, etc. Try to not do it manually.

```python
new_names = []
for nam in coffee_data.columns:
    new_name = nam
    new_name = new_name.strip().replace(" ", "_").lower()
    new_name = new_name.strip().replace(".", "_").lower()
    new_name = new_name.strip().replace(":", "_").lower()
    new_names.append(new_name)

coffee_data.columns = new_names
coffee_data.columns
```

### Problem 4. Bag weight (1 point)
What's up with the bag weights? Make all necessary changes to the column values. Don't forget to document your methods and assumptions.

```python
coffee_data["bag_weight"].unique()
```

```python
for idx, val in enumerate(coffee_data["bag_weight"]):
    values = val.split()
    num = values[0]
    
    if len(values) < 2:
        values.append('kg')  
    
    if len(values) >= 2:
        current_weight = values[1]
        if current_weight == 'lbs':
            converted_weight = float(num) * 0.45
            values[0] = f"{converted_weight:.2f}"  
            values[1] = 'kg'

        coffee_data.at[idx, 'bag_weight'] = " ".join(values)
```

```python
coffee_data.bag_weight.unique()
```

The values are either in kg or lbs or in both kg and lbs. The elements in lbs should be converted to kg and in case one has both measures - to keep both.
